## Detection of Road Edge Algorithm

Goal of Edge Detection: indentifying sharp changes in intensity in adjacent pixels.





#### Gradient: measure of change in brightness over adjacent pixels

- using brightness to evaluate gradient

#### Loading gray image:

```python
image = cv2.imread('test_image.jpg') #read the image from the file

lane_image = np.copy(image)
gray =cv2.cvtColor(lane_image, cv2.COLOR_RGB2GRAY)

#inshow() show the image with('title', and object)
cv2.imshow('result', gray) 

#displays the image for a specified amount of millisconds.
cv2.waitKey(0) 
```

#### Covert image to Grayscale and Reduce Noise



- Gaussian Blur

```python
image = cv2.imread('test_image.jpg') #read the image from the file

lane_image = np.copy(image)
gray =cv2.cvtColor(lane_image, cv2.COLOR_RGB2GRAY)


#this step is optional when using Canny method
blur = cv2.GaussianBlur(gray, (5,5), 0) #reduce Noise from our grayscale image

#inshow() show the image with('title', and object)
cv2.imshow('result', blur) 

#displays the image for a specified amount of millisconds.
cv2.waitKey(0) 
```

#### Canny Method: Edge Detection

A function of pixel intensities, in all directions X and y. - Derivative(f(x, y))
- small change, small derivative,
- large change, large derivative.

- cv2.Canny(image, low_threshold, high_threshold)

```python
image = cv2.imread('test_image.jpg') #read the image from the file

lane_image = np.copy(image)
gray =cv2.cvtColor(lane_image, cv2.COLOR_RGB2GRAY)


#this step is optional when using Canny method
blur = cv2.GaussianBlur(gray, (5,5), 0) #reduce Noise from our grayscale image

#apply canny in blur image
canny = cv2.Canny(blur, 50, 150) #low_threshold, high_threshold

#inshow() show the image with('title', and object)
cv2.imshow('result', canny) 

#displays the image for a specified amount of millisconds.
cv2.waitKey(0) 
```
